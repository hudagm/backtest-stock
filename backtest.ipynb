```python
import pandas as pd
import numpy as np
```

```python
## Generate random price history of stocks
def StockData(mean, std_dev):
    df = pd.DataFrame(np.random.normal(mean, std_dev, size = (100,1)), columns = ['price'])
    return df[df['price'] > 0].reset_index(drop = True)

stockdf = StockData(0, 0.1)
stockdf
```

```python
'''
Trading rules:
- if current units purchased > 0:
	(price today - price yesterday < -0.01), purchase 1 unit
	(price today - price yesterday > 0.11), sell everything

- if current units purchased = 0:
	(price today > price yesterday), buy 1 unit

vars:
current_unit
net_worth = [1,10,100,1000]
profit = sell_price - buy price
Profit % = 100 * (Profit / buy price)
'''

net_worth = 1
capital = 10
current_unit, profit, buy, i, j = 0,0,0,0,0

for j in range(100):
    while (i < stockdf.shape[0] - 1):
        if ((net_worth < 0) and (capital < 0)):
            print('Insuficient capital, cannot trade!')
            break
        if ((current_unit == 0) and (stockdf['price'][i+1] > stockdf['price'][i])):  #current unit == 0 --> buy 1 unit
            capital -= ((stockdf['price'].values[i+1]) * 1)
            buy += ((stockdf['price'].values[i+1]) * 1)
            current_unit += 1  
        elif current_unit > 0:
            if ((stockdf['price'][i+1] - stockdf['price'][i]) < -0.01):      # current unit < -0.01 --> buy 1 unit
                capital -= ((stockdf['price'].values[i+1]) * 1)  # buying 1 unit
                buy += ((stockdf['price'].values[i+1]) * 1)
                current_unit += 1
            elif ((stockdf['price'][i+1] - stockdf['price'][i]) > 0.11):    # current unit > 0.11 --> sell all
                profit = ((current_unit * (stockdf['price'].values[i+1])) - buy)
                capital += (current_unit * (stockdf['price'].values[i+1]))  
                current_unit = 0
            else:
                current_unit = current_unit
        else:
            current_unit = current_unit
        net_worth = capital + profit
        i += 1

print('Current Units:', current_unit)
print('Capital:', capital)
print('Profit:', profit)
print('Net Worth:', net_worth)
```
